#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
calificaciones=[8.5, 9.2, 7.8, 8.9, 10]
media_calificaciones = sum(calificaciones) / len(calificaciones)
print(media_calificaciones)
# otra forma mejor
from statistics import mean
print(mean(calificaciones))
```

```python
print(max(calificaciones))
print(min(calificaciones))
```

#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
max_price=ds['price'].max()
print(max_price)
address_max_price=ds[ds['price']== max_price]['address'].iloc[0]
print(address_max_price)
print(f'La casa con la dirección en {address_max_price} es la más cara y su precio es de {max_price}USD')
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000USD `

```python
ds=ds[ds['price']!=0]
#supongo que no regalan las casas, por lo que no tengo en cuenta las que no tienen precio
min_price=ds['price'].min()
print(min_price)
address_min_price=ds[ds['price']== min_price]['address'].iloc[0]
print(address_min_price)
print(f'La casa con la dirección en {address_min_price} es la más cara y su precio es de {min_price} USD')
#ese precio tan bajo indica que es el precio de alquiler no de venta, por lo que habría que filtrar sólo por el campo de en venta
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
precios_Arroyo=ds[ds['level5']=="Arroyomolinos (Madrid)"]
data=precios_Arroyo['price']
plt.figure(figsize = (10, 5))

plt.hist(data, bins = 30, alpha = 0.7)

plt.title("Histograma")
plt.show()
```
